# Pandas and the Titanic dataset



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Titanic dataset (if not already loaded)
df = pd.read_csv("https://gist.githubusercontent.com/teamtom/1af7b484954b2d4b7e981ea3e7a27f24/raw/114fb69dce56b4462a9c3a417e7402330616ad4f/titanic_full.csv")  # Update this with the correct file path

sns.set_style("whitegrid")  # Set Seaborn style
```

```python
# Survival Rate by Sex
plt.figure(figsize=(5, 4))
sns.barplot(x=df["sex"], y=df["survived"], palette="Set2")
plt.xlabel("Sex")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Sex")
plt.show();
```

```python
# Survival Rate by Passenger Class
plt.figure(figsize=(8, 5))
sns.barplot(x=df["pclass"], y=df["survived"], palette="coolwarm")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Passenger Class")
plt.show();
```

```python
# Survival Rate by Age Group
bins = [0, 9, 19, 29, 39, 49, 59, 120]
labels = ['0-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60+']
df["age_group"] = pd.cut(df["age"], bins=bins, labels=labels)

plt.figure(figsize=(10, 5))
sns.barplot(x=df["age_group"], y=df["survived"], palette="viridis", order=labels)
plt.xlabel("Age Group")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Age Group")
plt.show();
```

```python
# Survival Rate by Embarkation Port
plt.figure(figsize=(6, 4))
sns.barplot(x=df["embarked"], y=df["survived"], palette="Blues_d")
plt.xlabel("Embarked Port")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Embarkation Port")
plt.show();
```

```python
# Survival Rate by Family Size
df["family_size"] = df["sibsp"] + df["parch"]

plt.figure(figsize=(10, 5))
sns.lineplot(x=df["family_size"], y=df["survived"], marker="o", color="red")
plt.xlabel("Family Size")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Family Size")
plt.xticks(range(df["family_size"].max() + 1))
plt.show();
```

```python
# Wealth and Survival (Fare Analysis)
bins = [0, 10, 50, df["fare"].max()]
labels = ["Low Fare", "Medium Fare", "High Fare"]
df["fare_category"] = pd.cut(df["fare"], bins=bins, labels=labels)

plt.figure(figsize=(7, 4))
sns.barplot(x=df["fare_category"], y=df["survived"], palette="magma")
plt.xlabel("Fare Category")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Fare Category")
plt.show();
```

```python
# Male vs. Female Survival Rate by Class
plt.figure(figsize=(8, 5))
sns.barplot(x="pclass", y="survived", hue="sex", data=df, palette="coolwarm")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Gender and Class")
plt.legend(title="Sex")
plt.show();
```

## Summary of Visuals:
1. **Bar chart**: Survival rate by **sex**.
1. **Bar chart**: Survival rate by **class**.
2. **Bar chart**: Survival rate by **age group**.
3. **Bar chart**: Survival rate by **embarkation port**.
4. **Line plot**: Effect of **family size** on survival.
5. **Bar chart**: Impact of **fare paid** on survival.
6. **Grouped bar chart**: **Gender-class interaction** in survival.
